```python
import pandas as pd
```

```python
df=pd.read_csv('movies.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df=df[['id','title','genre','overview']]
```

```python
df.head()
```

```python
df["keyword"]=df['genre']+df['overview']
```

```python
df.head()
```

```python
df.drop(columns=['genre','overview'],inplace=True)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=9985,stop_words="english")
```

```python
vectors=cv.fit_transform(df["keyword"].values).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[2])),reverse=True,key=lambda x:x[1])[1:4]
```

```python
def Recommend(movie):
  movie_index=df[(df['title'])==(movie)].index[0]
  distance=sorted(list(enumerate(similarity[movie_index])),reverse=True,key=lambda x:x[1])[1:6]
  for i in distance:
    print(df.iloc[i[0]].title)
```

```python
Recommend("The Godfather")
```
